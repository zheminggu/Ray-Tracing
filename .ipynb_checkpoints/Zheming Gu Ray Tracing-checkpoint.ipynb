## help functions
these are some help funcitons to make calculation easier

<strong> lerp </strong> : liner interpolation, which can lerp color and normal in my program

<strong> normalize </strong> : normalize vectors

<strong> clamp </strong> : get a number between min number and max number, if smaller than min number return min number, if larger than max number return max number

<strong> update_progress </strong> : in render process, use this function to show progress


```python
def lerp(v1,v2,percentage):
    v1 = np.array(v1)
    v2 = np.array(v2)
    return v1+ percentage*(v2-v1)

def normalize(v):
    norm = np.linalg.norm(v)
    if norm == 0:
        return v
    return np.array(v / norm)

def clamp(number, min_number, max_number):
    return max(min_number, min(number, max_number))

def update_progress(progress):
    bar_length = 20
    if isinstance(progress, int):
        progress = float(progress)
    if not isinstance(progress, float):
        progress = 0
    if progress < 0:
        progress = 0
    if progress >= 1:
        progress = 1

    block = int(round(bar_length * progress))
    clear_output(wait = True)
    text = "Progress: [{0}] {1:.1f}%".format( "#" * block + "-" * (bar_length - block), progress * 100)
    print(text)
```

## From screen to viewport
in order to shoot a ray from screen to the actural scene, we need transfrom the coordinate from screen to viewport of camera, which is in the world coordinate 

```python
def screen_to_viewport(s_x, s_y):
    return np.array([s_x*v_w/screen_width, -s_y*v_y/screen_height, d ])
```

## shoot a ray to get color
after translate form scrren to viewport we could get a ray from camera position to a point in the viewport, trace this ray to get color of the scene, this part is the main function of ray tracing

```python
def closest_intersection(original, direction,min_len, max_len):
    closest_model={}
    z_buffer  = math.inf
    for model in models:
        if model["type"] == "sphere":
            x1,x2 = detect_intersect_sphere(original,direction,model)
            # if x1 != math.inf and x2 != math.inf:
            #     print(x1,x2)
            if x1 >= min_len and x1< max_len and x1 < z_buffer:
                z_buffer = x1
                closest_model = model
            if x2 >= min_len and x2< max_len and x2 < z_buffer:
                z_buffer = x2
                closest_model = model
    return closest_model, z_buffer


def get_reflect_ray(ray, normal):
    return 2 * normal * np.dot(normal , ray) - ray

def ray_tracing(original_position, v_ray,min_len,max_len, depth):
    color_ray_vector = (255 *(v_ray[0]+ 0.5),255 *(v_ray[1]+ 0.5), 255 * v_ray[2]) 
    color_ray = np.array([0,0,0])
    closest_model, z_buffer = closest_intersection(original_position, v_ray,min_len, max_len)
    
    if closest_model == {}:
        return color_ray
    intersect_position = original_position + z_buffer * v_ray
    normal = intersect_position - closest_model["center"]
    color_ray = calculate_color(closest_model, intersect_position,normal, v_ray, closest_model["specular"])
    # print(f"color_ray {color_ray}")
    r = closest_model["reflective"]
    if depth<=0 or r<=0:
        # print(color_ray)
        return  np.array(color_ray)
    R = get_reflect_ray(-v_ray, normal)
    reflected_color = ray_tracing(intersect_position, R, 0.001, math.inf, depth-1)
    # print(reflected_color)
    new_color = np.array(color_ray)*(1-r) + np.array(reflected_color)* r
    # print(f"new_color {new_color}")
    return new_color
```

## detect intersect
In order to get color of a ray , we need to judge if the ray is intersect with an object in a scene, if intersect then we can get the color from this object surface.
for now because all the object is sphere we can detect intersect in a simple way

```python
def detect_intersect_sphere(original_position, v_ray, sphere):
    sphere_center = np.array(sphere["center"])
    sphere_radius = sphere["radius"]
    v_camera_sphere = original_position - sphere_center
    a  = np.dot(v_ray,v_ray)
    b = 2 * np.dot(v_camera_sphere,v_ray)
    c = np.dot(v_camera_sphere,v_camera_sphere) -sphere_radius *sphere_radius
    delta = b*b -4*a*c
    if delta < 0:
        return math.inf, math.inf
    else:
        return (-b +math.sqrt(delta))/(2*a) , (-b - math.sqrt(delta))/ (2* a)
```

## calculate color
after light hit sphere calculate color


```python
def get_normal_from_texture(model_texture,position):
    # print(f"model is not none , previous normal {position}")
    color = get_color_from_texture("sphere", model_texture, position)
    normal =[0,0,0]
    normal[0]= 2.*color[0]/255.-1.
    normal[1]= 2.*color[1]/255.-1.
    normal[2]= -1.* color[2]/128.+2.
    # print(normal)
    normal = np.array(normal)+position
    # print(f"color{color} changed notmal {normal}")
    return normal


def get_color_from_texture(mapping_type, model_texture, position):
    position = normalize(position)
    if mapping_type =="sphere":
        u= 0.5 + math.atan2(position[2],position[0])/(2* math.pi)
        v = 0.5 - math.asin(position[1])/math.pi
        actual_u = u*(model_texture.shape[0]-1)
        actual_v = v*(model_texture.shape[1]-1)
    else:
        u= 0.5 + math.atan2(position[2],position[0])/(2* math.pi)
        v = 0.5 - math.asin(position[1])/math.pi
        actual_u = u*(model_texture.shape[0]-1)
        actual_v = v*(model_texture.shape[1]-1)
        # actual_u = position[0] % model_texture.shape[0]
        # actual_v = position[2] % model_texture.shape[1]
        # print(position)
    low_u = int(actual_u)
    high_u = low_u if low_u==model_texture.shape[0]-1 else low_u + 1
    low_v = int(actual_v)
    high_v = low_v if low_v==model_texture.shape[1]-1 else low_v+1
    # print(actual_u,actual_v)
    color1 = lerp(model_texture[low_u][low_v], model_texture[high_u][low_v], actual_u-low_u)
    color2 = lerp(model_texture[low_u][high_v], model_texture[high_u][high_v], actual_u-low_u)
    color_final = lerp(color1,color2,actual_v-low_v)
    return color_final

def calculate_color(model, intersect_position, normal, view_vector, specular):
    model_texture = model["texture"]
    model_color = model["color"]
    if type(model_texture) is not np.ndarray:
        model_texture =np.array( model_texture)
    if type(model_color) is not np.ndarray:
        model_color = np.array(model_color)
    # print(intersect_position)
    color = []
    if model["bump_mapping_texture"] is not None:
        normal = get_normal_from_texture(model["bump_mapping_texture"],normal)
    normal = normalize(normal)
    ambient_color =ambient["intensity"] * np.array(ambient["color"])
    color = ambient_color
    for light in lights:
        if light["type"]== "ambient":
            i += light["intensity"]
        else:
            light_dir = []
            if light["type"] == "point":
                light_dir = light["position"]-intersect_position
            elif light["type"] == "directional":
                light_dir = light["direction"]
            light_dir = normalize(light_dir)
            # shadow
            if light["produce_shadow"]:
                shadow_sphere, shadow_t = closest_intersection(intersect_position, light_dir, 0.001, math.inf)
                if shadow_sphere != {}:
                    # print("shadow here")
                    continue

            camera_dir = -view_vector
            H = np.array(light_dir+camera_dir)
            H = normalize(H)
            smoothfalloff = max(pow(np.dot(normal,H),specular),0)
            if model["render_type"]=="texture":
                if model["texture_mapping"] == "sphere":
                    diffuse_color = light["intensity"] * np.array(get_color_from_texture("sphere", model_texture,normal)) * np.dot(normal, light_dir)
                else:
                    diffuse_color = light["intensity"] * np.array(get_color_from_texture("color", model_texture,intersect_position)) * np.dot(normal, light_dir)
            else:
                diffuse_color = light["intensity"] * np.array(model_color) * np.dot(normal, light_dir)
            specular_color =  light["intensity"] * np.array(light["color"]) *smoothfalloff
            color += diffuse_color + specular_color

    for i in range(len(color)):
        color[i] = clamp(color[i],0,255)
    return np.array(color)
    # print(i)
```

## Show Images 
after tracing the ray, we could get the color from this specific screen position, we need to find a way to assign the color to the right position of image buffer
if you want to save image, uncomment the last line in this block

```python
def set_image_buffer(x,y, color):
    x = int(max(0, min(screen_width-1, x)))
    y = int(max(0, min(screen_height-1, y)))
    # print(x,y, color)
    if len(color) == 3:
        image_buffer[y][x][0] = color[0]
        image_buffer[y][x][1] = color[1]
        image_buffer[y][x][2] = color[2]
        image_buffer[y][x][3] = 255
    elif len(color) == 4:
        image_buffer[y][x] = tuple(color)

def show_image():
    image = Image.new('RGBA',(screen_width,screen_height), (0,0,0,255))
    d = ImageDraw.Draw(image)
    for h in range(screen_height):
        for w in range(screen_width):
            d.point((w,h),fill = (int(image_buffer[h][w][0]),int(image_buffer[h][w][1]),int(image_buffer[h][w][2]),255))
    image.show()
    image.save("ray tracing.png")
```

## ray tracing all the screen position
now we can ray tracing all the screen position with defined function 
for each pixel, shoot a ray from camera to viewport, tracing the ray color, and then write the color to image buffer 

```python
for x in range(int(-screen_width/2), int(screen_width/2)):
    for y in range(int(-screen_height/2), int(screen_height/2)):
        v_ray = screen_to_viewport(x, y)
        color = ray_tracing(camera_position, v_ray, min_len, max_len, ray_tracing_depth)
        set_image_buffer(x + screen_width/2,y + screen_height/2, color)
    update_progress((x+ screen_width/2)/float(screen_width))
update_progress(1)
show_image()
```
